This notebook rechunk the registered dataset using rechunker package, so that it's ready for g2 analysis. Time axis would not longer be chunked

```python
import hyperspy.api as hs
import numpy as np
import matplotlib.pyplot as plt
```

```python
from dask.distributed import Client, LocalCluster
cluster = LocalCluster(n_workers=4,
                      threads_per_worker=1,
                      dashboard_address=':8704')
cluster.scale(4)
client = Client(cluster)
```

```python
dc = hs.load('strip_reg_subpix.zspy',lazy=True)
```

```python
dc.data
```

```python
from rechunker import rechunk
```

```python
print(dc.data)
```

```python
rechunked = rechunk(dc.data,target_chunks=(4,4,500,15,160),max_mem='40GB',target_store='strip_10000fps_reg_rec.zarr',temp_store='temp.zarr')
```

```python
rechunked
```

```python
rechunked.execute()
```

```python
rechunked
```
